# 3. Statystyki opisowe

(przynajmniej: średnia, mediana, minimum, maksimum, odchylenie standardowe, skośność)

Statystyki opisowe dla danych przedstawiają się następująco:

* Liczba unikalnych wartości dla każdej zmiennej
* Najczęściej występująca wartość (top) oraz jej częstość (freq)

Niektóre zmienne mają tylko jedną unikalną wartość, np. veil-type, co oznacza, że są one stałe w całym zbiorze danych i mogą być potencjalnie usunięte z analizy, ponieważ nie wnoszą żadnej informacji.

### Obliczanie statystyk opisowych

```python
descriptive_stats = data.describe(include='all')
print("\nDescriptive Statistics:")
print(descriptive_stats)
```

### One Hot Recording

Tutaj zmienne kategoryczne zostana zamienione na zestaw zmiennych binarnych, gdzie każda kategoria jest reprezentowana jako oddzielna kolumna z wartościami 0 lub 1.

```python
data_encoded = pd.get_dummies(data)
print("\nData after One-Hot Encoding:")
print(data_encoded.head())
```

### Najcześćiej wystepujące wartości

```python
most_frequent_values = data.mode().iloc[0]
print("\nMost Frequent Values:")
print(most_frequent_values)
```

# 4. Podstawowa wizualizacja danych

(np. boxplot, histogram)

```python
plt.figure(figsize=(20, 15))
for i, column in enumerate(data.columns, 1):
    plt.subplot(6, 4, i)
    sns.countplot(y=data[column], order=data[column].value_counts().index)
    plt.title(column)
plt.tight_layout()
plt.show()
```

# 5. Transformacje danych

(skalowanie, logarytmowanie)

### Skalowanie damnych


Aby każda cecha miała średnia wartośc 0  i odchylenie standardowe 1 dokonaliśmy procesu skalowania naszych danych metodą standaryzacji, co w późniejszych etapach ułatwi dalsze obliczenia
$$
X _{scaled} = \frac{X−μ}{σ}
$$

* $X$ to wartość oryginalna,
* $μ$ to średnia wartość cechy,
* $σ$ to odchylenie standardowe cechy.


```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_encoded)
scaled_data = pd.DataFrame(scaled_data, columns=data_encoded.columns)

print("\nScaled Data:")
print(scaled_data.head())
```
